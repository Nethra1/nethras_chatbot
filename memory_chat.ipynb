```python
import os
from openai import OpenAI
from dotenv import load_dotenv
import gradio as gr
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
# gpt_key = os.getenv('OPENAI_API_KEY')
gemini_key = os.getenv('GOOGLE_API_KEY')

GEMINI_URL = "https://generativelanguage.googleapis.com/v1beta/"

gemini = OpenAI(base_url=GEMINI_URL, api_key=gemini_key)
```

```python
system_prompt_message = "You are a subject matter expert to answer below question and explain it as if you are explaining it to a 10 year old"
# user_prompt_message = "Can you tell me the difference between mcp and api in short?"
history = []

def generate_message(user):
    return [
    {
        "role":"system", "content":system_prompt_message
    }] + history + [{
        "role":"user","content":user
    }]
```

```python


def call_model_with_history(user_message):
    message = generate_message(user_message)

    response = gemini.chat.completions.create(model="gemini-2.5-pro", messages=message)
    display(Markdown(response.choices[0].message.content))
    reply = response.choices[0].message.content
    history.append({"role":"user", "content": user_message})
    history.append({"role":"assistant", "content": reply})
```

```python
user_message = "Tell me my name and today's date"
call_model_with_history(user_message)
```
