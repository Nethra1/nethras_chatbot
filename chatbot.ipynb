```python
import os
from openai import OpenAI
from dotenv import load_dotenv
import gradio as gr
```

```python
load_dotenv(override=True)
# gpt_key = os.getenv('OPENAI_API_KEY')
gemini_key = os.getenv('GOOGLE_API_KEY')

GEMINI_URL = "https://generativelanguage.googleapis.com/v1beta/"

gemini = OpenAI(base_url=GEMINI_URL, api_key=gemini_key)
```

```python
system_prompt_message = "You are an adaptive Subject Matter Expert (SME) assistant.\n\nYour role:\n- You are an expert in *whatever subject the user asks about* — technical, scientific, business, creative, or academic.\n- Identify the topic, adopt the role of an experienced expert, and answer with authority.\n- Explain concepts **as if teaching a beginner** in that subject — simple, clear, and step-by-step.\n- Keep responses **sharp, short, and crisp**, without unnecessary jargon.\n- Use examples or analogies only when they make understanding easier.\n- If the question is unclear, ask for clarification before answering.\n\nYour goal:\nDeliver accurate, beginner-friendly, and expert-level explanations that are concise and easy to understand."
# user_prompt_message = "Can you tell me the difference between mcp and api in short?"
history = []

def generate_message(user, history):
    return [
    {
        "role":"system", "content":system_prompt_message
    }] + history + [{
        "role":"user","content":user
    }]
```

```python
def chat(message, history):
    print(history)
    history = [{"role":hist["role"], "content": hist["content"]} for hist in history]
    message = generate_message(message, history)
    response = gemini.chat.completions.create(model="gemini-2.5-pro", messages=message)
    # display(Markdown(response.choices[0].message.content))
    reply = response.choices[0].message.content
    return reply
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```
